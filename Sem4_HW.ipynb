```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('kc-house-data.csv', sep = ',')
df.head()
```

1.2 Изучите стоимости недвижимости

```python
plt.figure(figsize=(3, 3))
plt.hist(df['price'], bins=8)
plt.title('Price of houses')
plt.xlabel('Price')
plt.ylabel('Quantity');
```

1.3. Изучите распределение квадратуры жилой

```python
plt.figure(figsize=(3,3))
plt.hist(df['sqft_living'], bins=20);
```

1.4. Изучите распределение года постройки

```python
data = df['yr_built'].value_counts()
plt.scatter(data.index, data.values);
```

2.1. Изучите распределение домов от наличия вида на набережную. Постройте график

```python
data = df['waterfront'].value_counts()
data.index = ['no', 'yes']
plt.figure(figsize=(3, 3))
plt.pie(data, autopct='%1.1f%%')
plt.title('Percents of waterfront flats')
plt.legend(data.index);
```

2.2. Изучите распределение этажей домов

```python
data = df['floors'].value_counts()
plt.figure(figsize=(5, 3))
plt.bar(data.index, data.values, width=0.4)
plt.xlabel('Floors')
plt.ylabel('Quantity')
plt.title('Floors chart');
```

2.3. Изучите распределение состояния домов

```python
data = df['condition'].value_counts()
plt.figure(figsize=(5, 3))
plt.bar(data.index, data.values)
plt.xlabel('Condition')
plt.ylabel('Quantity')
plt.title('Condition chart');
```
